```python
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW

import pandas as pd
import numpy as np
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve, classification_report
from tqdm import tqdm
import os
import glob # For finding file paths

from einops import rearrange
from einops.layers.torch import Rearrange
import albumentations as A
from albumentations.pytorch import ToTensorV2

# Set style for all plots
sns.set_style("whitegrid")

# ==============================================================================
# 1. EDA AND DATA PREPROCESSING
# ==============================================================================

def show_image_grid(df, labels, n_images=25):
    """Visualizes a 5x5 grid of sample images with their labels."""
    plt.figure(figsize=(15, 15))
    sample_df = df.sample(n_images)
    
    for i, (idx, row) in enumerate(sample_df.iterrows()):
        plt.subplot(5, 5, i + 1)
        image_path = row['FullPath']
        image = cv2.imread(image_path)
        
        if image is None:
            print(f"Warning: Could not load image at {image_path}. Skipping.")
            continue
            
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        image_labels = [label for label in labels if row[label] == 1]
        title = "\n".join(image_labels) if image_labels else "No Finding"
        
        plt.imshow(image)
        plt.title(title, fontsize=8)
        plt.axis('off')
        
    plt.tight_layout()
    plt.show()

class NIHChestXrayDataset(Dataset):
    """Custom Dataset for NIH Chest X-ray images."""
    def __init__(self, df, labels, transform=None):
        self.df = df
        self.transform = transform
        self.labels = labels

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image_path = row['FullPath']
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        
        if self.transform:
            image = self.transform(image=image)['image']
        
        labels_vector = torch.tensor(row[self.labels].values.astype(np.float32))
        return image, labels_vector

def get_transforms(image_size):
    """Returns training and validation augmentations."""
    mean, std = [0.485, 0.456, 0.406], [0.229, 0.224, 0.225]
    
    train_transform = A.Compose([
        A.Resize(image_size, image_size),
        A.HorizontalFlip(p=0.5),
        A.Rotate(limit=15, p=0.5),
        A.Normalize(mean=mean, std=std),
        ToTensorV2()
    ])
    
    val_transform = A.Compose([
        A.Resize(image_size, image_size),
        A.Normalize(mean=mean, std=std),
        ToTensorV2()
    ])
    return train_transform, val_transform

# ==============================================================================
# 2. MODEL DEFINITION: SemaCheX-Former
# ==============================================================================
class TransformerEncoderBlock(nn.Module):
    def __init__(self, dim, num_heads, mlp_dim, dropout=0.1):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn = nn.MultiheadAttention(dim, num_heads, dropout=dropout, batch_first=True)
        self.norm2 = nn.LayerNorm(dim)
        self.mlp = nn.Sequential(nn.Linear(dim, mlp_dim), nn.GELU(), nn.Dropout(dropout), nn.Linear(mlp_dim, dim), nn.Dropout(dropout))

    def forward(self, x):
        normed_x = self.norm1(x)
        attn_output, _ = self.attn(normed_x, normed_x, normed_x)
        x = x + attn_output
        x = x + self.mlp(self.norm2(x))
        return x

class PartitionReconstructionAttentionBlock_LMSA(nn.Module):
    def __init__(self, in_channels, dim, patch_size, num_heads, mlp_dim, num_classes, dropout=0.1):
        super().__init__()
        self.patch_size, patch_dim = patch_size, in_channels * patch_size * patch_size
        self.partition = Rearrange('b c (h p1) (w p2) -> b (h w) (p1 p2 c)', p1=patch_size, p2=patch_size)
        self.patch_projection = nn.Linear(patch_dim, dim)
        self.label_embeddings = nn.Parameter(torch.randn(1, num_classes, dim))
        self.attention_module = TransformerEncoderBlock(dim, num_heads, mlp_dim, dropout)

    def forward(self, x):
        b, _, h, w = x.shape
        tokens = self.partition(x)
        tokens = self.patch_projection(tokens)
        label_embs = self.label_embeddings.expand(b, -1, -1)
        combined_tokens = torch.cat((tokens, label_embs), dim=1)
        processed_tokens = self.attention_module(combined_tokens)
        processed_visual_tokens = processed_tokens[:, :tokens.shape[1]]
        recon_h, recon_w = h // self.patch_size, w // self.patch_size
        return rearrange(processed_visual_tokens, 'b (h w) d -> b d h w', h=recon_h, w=recon_w)

class ConvSEBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, reduction=8):
        super().__init__()
        self.conv_block = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size, padding='same', bias=False), nn.BatchNorm2d(out_channels), nn.ReLU(inplace=True))
        self.se = nn.Sequential(nn.AdaptiveAvgPool2d(1), nn.Conv2d(out_channels, out_channels // reduction, 1, bias=False), nn.ReLU(inplace=True), nn.Conv2d(out_channels // reduction, out_channels, 1, bias=False), nn.Sigmoid())

    def forward(self, x):
        cnn_out = self.conv_block(x)
        return cnn_out * self.se(cnn_out)

class SemaCheXFormer(nn.Module):
    def __init__(self, image_size=224, in_channels=3, num_classes=14, dim=256, num_heads=8, mlp_dim=512, cnn_channels=256):
        super().__init__()
        self.stem = nn.Sequential(nn.Conv2d(in_channels, 64, kernel_size=7, stride=2, padding=3, bias=False), nn.BatchNorm2d(64), nn.ReLU(inplace=True), nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
        self.attention_stage = PartitionReconstructionAttentionBlock_LMSA(64, dim, 7, num_heads, mlp_dim, num_classes)
        self.cnn_se_stage = ConvSEBlock(dim, cnn_channels)
        self.classification_head = nn.Sequential(nn.AdaptiveAvgPool2d(1), nn.Flatten(), nn.Linear(cnn_channels, num_classes))

    def forward(self, x):
        return self.classification_head(self.cnn_se_stage(self.attention_stage(self.stem(x))))


# ==============================================================================
# 3. TRAINING, EVALUATION, AND PLOTTING FUNCTIONS
# ==============================================================================

def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for images, labels in tqdm(dataloader, desc="Training"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        all_preds.append(torch.sigmoid(outputs).detach().cpu().numpy())
        all_labels.append(labels.cpu().numpy())
        
    avg_loss = total_loss / len(dataloader)
    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)
    accuracy = ( (all_preds > 0.5).astype(int) == all_labels ).mean()
    return avg_loss, accuracy

def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    all_preds, all_labels = [], []

    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc="Validating"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            all_preds.append(torch.sigmoid(outputs).cpu().numpy())
            all_labels.append(labels.cpu().numpy())

    avg_loss = total_loss / len(dataloader)
    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)
    accuracy = ( (all_preds > 0.5).astype(int) == all_labels ).mean()
    return avg_loss, accuracy, all_preds, all_labels

def plot_history(history):
    """Plots fancy loss and accuracy curves."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    
    ax1.plot(history['train_loss'], 'o--', label='Train Loss', color='deepskyblue')
    ax1.plot(history['val_loss'], 'o-', label='Validation Loss', color='royalblue')
    ax1.set_title('Model Loss Over Epochs', fontsize=16)
    ax1.set_xlabel('Epochs', fontsize=12)
    ax1.set_ylabel('Loss', fontsize=12)
    ax1.legend()
    
    ax2.plot(history['train_acc'], 'o--', label='Train Accuracy', color='lightcoral')
    ax2.plot(history['val_acc'], 'o-', label='Validation Accuracy', color='firebrick')
    ax2.set_title('Model Accuracy Over Epochs', fontsize=16)
    ax2.set_xlabel('Epochs', fontsize=12)
    ax2.set_ylabel('Accuracy', fontsize=12)
    ax2.legend()
    
    plt.show()

# <<<--- MODIFIED FUNCTION ---<<<
def plot_roc_curves(y_true, y_pred, labels):
    """Plots class-wise ROC AUC curves robustly."""
    plt.figure(figsize=(12, 10))
    for i, label in enumerate(labels):
        # Check if both positive and negative classes are present in the true labels
        if len(np.unique(y_true[:, i])) < 2:
            print(f"Skipping ROC curve for '{label}' because it only has one class in the validation set.")
            continue
        
        # If safe, calculate and plot
        fpr, tpr, _ = roc_curve(y_true[:, i], y_pred[:, i])
        auc = roc_auc_score(y_true[:, i], y_pred[:, i])
        plt.plot(fpr, tpr, label=f'{label} (AUC = {auc:.2f})')
        
    plt.plot([0, 1], [0, 1], 'k--', label='Chance')
    plt.title('Class-wise ROC AUC Curves', fontsize=16)
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.legend(loc='lower right')
    plt.show()

# <<<--- MODIFIED FUNCTION ---<<<
def print_and_plot_metrics(y_true, y_pred, labels):
    """Prints a consolidated report and plots a heatmap robustly."""
    y_pred_binary = (y_pred > 0.5).astype(int)
    
    print("üìã **Classification Report**\n" + "="*70)
    # This report is generally robust to single-class issues, but good practice to be aware.
    report = classification_report(y_true, y_pred_binary, target_names=labels, zero_division=0)
    print(report)
    
    # --- Robust AUC Score Calculation ---
    auc_scores = []
    for i, label in enumerate(labels):
        try:
            # Try to calculate the AUC score
            score = roc_auc_score(y_true[:, i], y_pred[:, i])
            auc_scores.append(score)
        except ValueError:
            # If it fails (due to one class present), append NaN and warn the user
            auc_scores.append(np.nan)
            print(f"‚ö†Ô∏è WARNING: ROC AUC score not defined for '{label}' (only one class present in y_true).")

    # Use nanmean to calculate the mean, ignoring any NaN values
    mean_auc = np.nanmean(auc_scores)
    
    print("\nüéØ **ROC AUC Scores**\n" + "="*70)
    for i, label in enumerate(labels):
        # Print the score, handling the display of NaN if it occurred
        if np.isnan(auc_scores[i]):
            print(f"{label:<20}: Not Defined")
        else:
            print(f"{label:<20}: {auc_scores[i]:.4f}")
    print(f"\n{'MEAN AUC (of defined)':<20}: {mean_auc:.4f}\n")
    
    # --- Plotting Heatmap ---
    report_dict = classification_report(y_true, y_pred_binary, target_names=labels, output_dict=True, zero_division=0)
    report_df = pd.DataFrame(report_dict).iloc[:-1, :].T
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(report_df, annot=True, cmap='viridis', fmt='.2f')
    plt.title('Classification Metrics Heatmap', fontsize=16)
    plt.show()


# ==============================================================================
# 4. MAIN EXECUTION SCRIPT
# ==============================================================================
if __name__ == '__main__':
    CONFIG = {
        "IMAGE_SIZE": 224, "BATCH_SIZE": 64, "LEARNING_RATE": 1e-4, "EPOCHS": 25,
        "DEVICE": "cuda" if torch.cuda.is_available() else "cpu",
        "DATASET_DIR": "/kaggle/input/data/", 
        "CSV_PATH": "/kaggle/input/data/Data_Entry_2017.csv"
    }
    
    print(f"Using device: {CONFIG['DEVICE']}")
    
    df = pd.read_csv(CONFIG['CSV_PATH'])

    print("Scanning for all image paths...")
    all_image_paths = glob.glob(os.path.join(CONFIG['DATASET_DIR'], '**', '*.png'), recursive=True)
    image_path_map = {os.path.basename(p): p for p in all_image_paths}
    
    df['FullPath'] = df['Image Index'].map(image_path_map)
    
    original_rows = len(df)
    df.dropna(subset=['FullPath'], inplace=True)
    print(f"Located {len(df)} of {original_rows} images listed in the CSV.")
    
    all_labels_str = df['Finding Labels'].str.get_dummies(sep='|')
    LABELS = [label for label in all_labels_str.columns if label != 'No Finding']
    df = pd.concat([df, all_labels_str[LABELS]], axis=1)
    
    print("Visualizing a 5x5 grid of sample images...")
    show_image_grid(df, LABELS)

    df_subset = df.sample(1000, random_state=42)
    train_df = df_subset.sample(frac=0.8, random_state=42)
    val_df = df_subset.drop(train_df.index)

    train_transform, val_transform = get_transforms(CONFIG["IMAGE_SIZE"])
    train_dataset = NIHChestXrayDataset(train_df, LABELS, transform=train_transform)
    val_dataset = NIHChestXrayDataset(val_df, LABELS, transform=val_transform)
    train_loader = DataLoader(train_dataset, batch_size=CONFIG['BATCH_SIZE'], shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=CONFIG['BATCH_SIZE'], shuffle=False, num_workers=2)

    model = SemaCheXFormer(num_classes=len(LABELS)).to(CONFIG['DEVICE'])
    criterion = nn.BCEWithLogitsLoss()
    optimizer = AdamW(model.parameters(), lr=CONFIG['LEARNING_RATE'])

    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    for epoch in range(CONFIG['EPOCHS']):
        print(f"\n--- Epoch {epoch+1}/{CONFIG['EPOCHS']} ---")
        train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, CONFIG['DEVICE'])
        val_loss, val_acc, y_pred, y_true = evaluate(model, val_loader, criterion, CONFIG['DEVICE'])
        
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        
        print(f"Epoch {epoch+1} | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

    print("\n\n--- üìà Final Evaluation and Visualization üìâ ---")
    plot_history(history)
    plot_roc_curves(y_true, y_pred, LABELS)
    print_and_plot_metrics(y_true, y_pred, LABELS)
```
